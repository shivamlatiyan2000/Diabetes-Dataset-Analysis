```python
#separating the dependent data and independent data
X=df.drop('Outcome',axis=1)
X
```

```python
Y=df['Outcome']
Y
```

```python
#feature scaling
#we should apply min max scaler
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
X_new=mn.fit_transform(X)
```

```python
X_new
```

```python
#divide it into 70:30
#training and testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_new,Y,test_size=0.3)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=23,metric='minkowski',weights='uniform')
knn.fit(X_train,Y_train)
```

```python
Y_Predict=knn.predict(X_test)
Y_Predict
```

```python
Y_test
```

```python
#in order to check the performance of classification model
#confusion matrix and classification report
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,Y_Predict))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_Predict))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(knn,X_train,Y_train,cv=7).mean()
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(knn,X_test,Y_test,cv=7).mean()
```

```python
#in order to find the best value of K
x=[]
scores=[]
for k in range(1,35):
    knn=KNeighborsClassifier(n_neighbors=k,metric='minkowski',weights='uniform')
    score=cross_val_score(knn,X_train,Y_train,cv=5)
    x.append(k)
    scores.append(score.mean())
```

```python
plt.plot(x,scores)
plt.xlabel('K Neighbors values')
plt.ylabel('Accuracy Scores')
plt.grid()
plt.figure(figsize=(2,2))
plt.show()
```

```python
#in order to find the best value of K
xt=[]
st=[]
for k in range(1,35):
    knn=KNeighborsClassifier(n_neighbors=k,metric='minkowski',weights='uniform')
    score=cross_val_score(knn,X_test,Y_test,cv=5)
    xt.append(k)
    st.append(score.mean())
```

```python
plt.plot(xt,st)
plt.xlabel('K Neighbors values')
plt.ylabel('Accuracy Scores')
plt.grid()
plt.show()
```
